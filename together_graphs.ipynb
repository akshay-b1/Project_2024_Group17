```python
algorithm_input_type_mapping = {
    'bitonic': {0: 'Sorted', 1: 'Sorted with 1% perturbed', 2: 'Reverse Sorted', 3: 'Random'},
    'radix_sort': {0: 'Sorted', 1: 'Sorted with 1% perturbed', 2: 'Reverse Sorted', 3: 'Random'},
    'merge_sort': {'Random': 'Random', 'Sorted': 'Sorted', 'ReverseSorted': 'Reverse Sorted', '1_perc_perturbed': 'Sorted with 1% perturbed'},
}


def update_input_type(row):
    if row['algorithm'] in algorithm_input_type_mapping:
        row['input_type'] = algorithm_input_type_mapping[row['algorithm']][row['input_type']]
    return row

tk.dataframe = tk.dataframe.reset_index().apply(update_input_type, axis=1).set_index(["algorithm", "node", "num_procs", "input_size", "input_type"]).sort_index()
```

```python
import thicket as th
from glob import glob

import matplotlib.pyplot as plt
import pandas as pd
num_procs = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]
input_sizes = [2**i for i in range(22, 23, 2)]
input_types = ["Random"]
algorithms = ['radix_sort', 'merge', 'bitonic', 'sample_sort']
directories = {'radix_sort': "radix_sort/radixsortcalifiles", "merge": "merge_sort/cali_files", "bitonic": "BitonicSort/cali_files", "sample_sort": "samplesort/samplesortcalifiles"}


pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

def create_scaling_plots():
    nodes = ['comm']
    input_sizes = [2**i for i in range(22, 23, 2)]
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.set_title(f'Strong Scaling - {nodes[0]} - Input Size {input_sizes[0]}', fontsize=16)
    algorithms = ['radix_sort', 'merge', 'bitonic', 'sample_sort']
    
    for algorithm in algorithms:
        df = temp[algorithm]
        alg_data = df.loc[algorithm]
        for node in nodes:
            for size in input_sizes:
                size_data = alg_data[alg_data.index.get_level_values('input_size') == size]
                for input_type in input_types:
                    input_data = size_data[size_data.index.get_level_values('input_type') == input_type]
                    tempList = [n.frame.get("name") for n in input_data.index.get_level_values("node").values]

                    nodeLoc = input_data.index.get_level_values("node").values[tempList.index(node)]
                    data = input_data.loc[nodeLoc]
                    
                    f = data['Total time'].iloc[0] / data['Total time']

                    if len(num_procs) != len(f):
                        wall_clock_times = data['Total time'] / data.index.get_level_values('num_procs')

                        base_time = wall_clock_times.iloc[0]
                        
                        speedups = base_time / wall_clock_times
                        ax.plot(num_procs[:-1], speedups, label=algorithm + ' ' + input_type)
                    else:
                        wall_clock_times = data['Total time'] / data.index.get_level_values('num_procs')

                        base_time = wall_clock_times.iloc[0]
                        
                        speedups = base_time / wall_clock_times
                        # speedup = 1 / (f + (1 - f) / num_procs)
                        # speedup = data['Avg time/rank'].iloc[0] / data['Total time']
                        ax.plot(num_procs, speedups, label=algorithm + ' ' + input_type)
                    

    ax.plot(num_procs, num_procs, linestyle='--', color='gray', label='Ideal')
    ax.set_xscale('log', base=2)
    ax.set_yscale('log', base=2)  
    ax.set_xlabel('Number of Processes')
    ax.set_ylabel('Speedup')
    # ax.set_ylim(0, 200)
    ax.legend(title='Input Type')
    plt.savefig(f'strong_scaling_{node}_input_size_{size}_input_type_{input_type[0]}.png')
    plt.close()

    nodes = ['comp_large']
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.set_title(f'Avg time/rank - {nodes[0]} - Input Size {input_sizes[0]}', fontsize=16)
    algorithms = ['radix_sort', 'merge', 'bitonic', 'sample_sort']

    for algorithm in algorithms:



        df = temp[algorithm]
        alg_data = df.loc[algorithm]
        for node in nodes:
            for size in input_sizes:
                size_data = alg_data[alg_data.index.get_level_values('input_size') == size]
                for input_type in input_types:
                    input_data = size_data[size_data.index.get_level_values('input_type') == input_type]
                    tempList = [n.frame.get("name") for n in input_data.index.get_level_values("node").values]

                    nodeLoc = input_data.index.get_level_values("node").values[tempList.index(node)]
                    data = input_data.loc[nodeLoc]
                    
                    f = data['Total time'].iloc[0] / data['Total time']

                    if len(num_procs) != len(f):
                        totalTime = data['Avg time/rank']
                        ax.plot(num_procs[:-1], totalTime, label=algorithm + ' ' + input_type)
                    else:
                        totalTime = data['Avg time/rank']
                        ax.plot(num_procs, totalTime, label=algorithm + ' ' + input_type)
                    
    ax.set_xscale('log', base=2)
    ax.set_xlabel('Number of Processes')
    ax.set_ylabel('Avg time/rank')
    ax.legend(title='Input Type')
    plt.savefig(f'avg_time_rank_{node}_input_size_{size}_input_type_{input_type[0]}.png')
    plt.close()

    nodes = ['comm']
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.set_title(f'Weak Scaling - {nodes[0]} - Input Size {input_sizes[0]}', fontsize=16)
    algorithms = ['radix_sort', 'merge', 'bitonic', 'sample_sort']
    input_sizes = [2**i for i in range(16, 30, 2)]
    for algorithm in algorithms:

        df = temp[algorithm]
        alg_data = df.loc[algorithm]
        for node in nodes:
            for i in [1]:
                proc_data = alg_data[alg_data.index.get_level_values('num_procs') == 128]
                for input_type in input_types:
                    input_data = proc_data[proc_data.index.get_level_values('input_type') == input_type]
                    tempList = [n.frame.get("name") for n in input_data.index.get_level_values("node").values]

                    nodeLoc = input_data.index.get_level_values("node").values[tempList.index(node)]
                    data = input_data.loc[nodeLoc]
                    totalTime = data['Total time']
                    if len(input_sizes) != len(totalTime):

                        
                        print(len(input_sizes), len(totalTime))
                        ax.plot(input_sizes[:-1], totalTime, label=algorithm + ' ' + input_type)
                    else:
                        print(len(input_sizes), len(totalTime))
                        ax.plot(input_sizes, totalTime, label=algorithm + ' ' + input_type)
                    

    ax.set_xscale('log', base=2)
    ax.set_xlabel('Input Size')
    ax.set_ylabel('Total Time')
    ax.legend(title='Input Type')
    plt.savefig(f'weak_scaling_{node}_input_size_{size}_input_type_{input_type[0]}.png')
    plt.close()


create_scaling_plots()

```
